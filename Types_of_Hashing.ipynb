# Hashing

```python
import hashlib

md5 = hashlib.md5(b'Cryptography')
sha256 = hashlib.sha256(b'Cryptography')
sha512 = hashlib.sha512(b'Cryptography')
sha1 = hashlib.sha1(b'Cryptography')

print(md5.hexdigest())
print(sha256.hexdigest())
print(sha512.hexdigest())
print(sha1.hexdigest())
```

```python
#convert all passwords to hash and store them in these four different dictionary.
import hashlib
file_name = '/content/passlist.txt'
password_dictionary_sha256 = {}
password_dictionary_sha512 = {}
password_dictionary_md5 = {}
password_dictionary_sha1 = {}

#todo
with open(file_name, "r") as f:
    for line in f:
        password = line.strip()
        pw_bytes = password.encode("utf-8")

        md5_hash = hashlib.md5(pw_bytes).hexdigest()
        sha1_hash = hashlib.sha1(pw_bytes).hexdigest()
        sha256_hash = hashlib.sha256(pw_bytes).hexdigest()
        sha512_hash = hashlib.sha512(pw_bytes).hexdigest()

        password_dictionary_md5[password] = md5_hash
        password_dictionary_sha1[password] = sha1_hash
        password_dictionary_sha256[password] = sha256_hash
        password_dictionary_sha512[password] = sha512_hash


print(password_dictionary_sha256)
print(password_dictionary_sha512)
print(password_dictionary_sha1)
print(password_dictionary_md5)
```

```python
#Break the password

#todo
admin_hash = "8062dc84e0faa9e553192841432d4836208f700d43f3b68a53cc30ad4b7b3464"
found_password = None
found_type = None

for password, hashed in password_dictionary_sha256.items():
    if hashed == admin_hash:
        found_password = password
        found_type = "SHA-256"
        break

if found_password is None:
    for password, hashed in password_dictionary_sha512.items():
        if hashed == admin_hash:
            found_password = password
            found_type = "SHA-512"
            break

if found_password is None:
    for password, hashed in password_dictionary_sha1.items():
        if hashed == admin_hash:
            found_password = password
            found_type = "SHA-1"
            break


if found_password is None:
    for password, hashed in password_dictionary_md5.items():
        if hashed == admin_hash:
            found_password = password
            found_type = "MD5"
            break


print(f"Password found: '{found_password}' (Hash type: {found_type})")


```
