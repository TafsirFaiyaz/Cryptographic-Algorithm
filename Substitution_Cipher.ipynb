# Caesar Cipher

```python
def decrypt_shift_cipher(ciphertext, key):
    plain_text = ""

    for char in ciphertext:
        if char == " ":
            plain_text += " "
            continue

        if char.isupper():
            # Shift within 'A'-'Z'
            plain_text += chr((ord(char) - 65 - key) % 26 + 65)
        elif char.islower():
            # Shift within 'a'-'z'
            plain_text += chr((ord(char) - 97 - key) % 26 + 97)
        else:
            # Keep punctuation/symbols unchanged
            plain_text += char

    return plain_text
```

```python
import matplotlib.pyplot as plt

def Frequency_Table_BarChart(text):
    text = text.lower()
    count = sum(1 for i in text if i in 'abcdefghijklmnopqrstuvwxyz')
    table = []
    for i in range(97, 123):
        char = chr(i)
        freq = text.count(char) / count if count > 0 else 0
        table.append((char, freq))
    table.sort(key=lambda x: x[1], reverse=True)
    for char, freq in table:
        print(f"{char}: {freq:.4f}")
    letters = [item[0] for item in table]
    frequencies = [item[1] for item in table]

    plt.figure(figsize=(11, 5))
    bars = plt.bar(letters, frequencies, width=0.6, color='skyblue')
    for bar, freq in zip(bars, frequencies):
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, height + 0.002,
                 f"{freq:.3f}", ha='center', va='bottom', fontsize=8)
    plt.title("Letter Frequency in Ciphertext")
    plt.xlabel("Letters")
    plt.ylabel("Frequency")
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
```
