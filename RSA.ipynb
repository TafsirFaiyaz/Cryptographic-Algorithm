```python
import random
import sympy

low = 1 << 127
high = (1 << 128) - 1

p = sympy.randprime(low, high)
q = sympy.randprime(low, high)

while q == p:
    q = sympy.randprime(low, high)

x = sympy.isprime(p)
y = sympy.isprime(q)

print("p:", p, )
print(f"Prime status:", x)
print("q:", q)
print(f"Prime status:",y)
print("bit-length p:", p.bit_length(), "bit-length q:", q.bit_length())
```

```python
#Calculate n = pq

n = p*q
print(n)
```

```python
#Calculate phi(n)=(p-1)(q-1)
phi_n = (p-1)*(q-1)
print(phi_n)
```

```python
#Perform encryption here and generate ciphertext [c = m^e (mod n)] and private key [d*e â‰¡ 1 (mod(phi(n)))]

import math

def choose_e(phi_n):
    e = 3
    while math.gcd(e, phi_n) != 1:
        e += 2
    return e

def egcd(a, b):
    if b == 0:
        return a, 1, 0
    g, x1, y1 = egcd(b, a % b)
    return g, y1, x1 - (a // b) * y1

def modinv(a, m):
    g, x, _ = egcd(a, m)
    if g != 1:
        raise ValueError("No inverse for chosen e")
    return x % m

e = choose_e(phi_n)

d = modinv(e, phi_n)

c = pow(m, e, n)



print("e:", e)
print("d:", d)
print("ciphertext:", c)

```

```python
#perform decryption here [m = c^d (mod n)]
m = pow(c, d, n)
```
