```python
import os
import sys
import binascii

from cryptography.hazmat.primitives import cmac
from cryptography.hazmat.decrepit.ciphers import algorithms

messages = [
    "I met an interesting turtle while the song on the radio blasted away",
    "I like to leave work after my eight-hour tea-break",
    "Her daily goal was to improve on yesterday",
    "He found the chocolate covered roaches quite tasty",
    "After fighting off the alligator, Brian still had to face the anaconda",
    "He decided to count all the sand on the beach as a hobby",
    "The sign said there was road work ahead so he decided to speed up",
    "Send 500$ to this account - 6589415651548",
    "Garlic ice-cream was her favorite",
    "I'd rather be a bird than a fish"
]

keys = [
    b'\x01\xd8i\xa1^0\x9a<\x0f\xf0\r\xc1\xdd\xd5\x89\xa6',
    b'\xa6+\x16\x9d-1\xda\x8aV\xed\xf5\xf0cv\x04\x88',
    b'[\xc5\xbd\xe4z\xd1=E\x17-ku\x02=|=',
    b'5"k\xff\x81a\x9b 7\x8c>\xb7\xb9\xdcu\xaa',
    b'\xa1\xfcw"?3\x91\x1c\t\x9c\x91\xe2He\x935',
    b'\xa7\x83@\xde\xbf\xb494\xee\x84\x1e-\xc8A\xf9:',
    b'2\xcbv\xdcU6\x99\xb6.\xa7\xea\xeb\xaf\x10\xc7\x90',
    b'\xc3\xea\x99e\xaal\xab\xd4\x9b\xf9\xb4Z\x19\xed\xcf\xcb',
    b'\x05\xf9\x83\x9d\xb7\xb6\xc3\xb8\x9e\xc5\xd9\xd8\x07]\xc6\xb3',
    b'\x84YY\xf0\x02GU\xa4LD\xd5\x85!A\xc2c'
]

mac_signatures = [
    "ba4ecb8db45c6ae0",
    "f47e78c537fa1435",
    "ddaf3152edbe868a",
    "9d30d856f84489a8",
    "b9d173e05bbf7738",
    "6355e471bd9930a1",
    "9fbafc75e0a5056a",
    "35273149636aca35",
    "dc2de1e07b71d391",
    "5e191d02aa5fc0b1"
]

for i in range(len(messages)):
    message = messages[i]
    message_bytes = message.encode()  # Convert string to bytes
    key = keys[i]
    # You code here. HINT: use CMAC, TripleDES
    c = cmac.CMAC(algorithms.TripleDES(key))

    c.update(message_bytes)

    signature = c.finalize()

    print(f"Message: {message}")
    print(f"CMAC signature: {binascii.b2a_hex(signature).decode()}")

    print("======================")
    given_signature = mac_signatures[i]
    if signature.hex() == given_signature:
        print("Valid")
    else:
        print("Invalid")

    print("********************************")
    print()
```

```python
def A51(X, Y, Z, n):
    X = list(X)
    Y = list(Y)
    Z = list(Z)

    keystream = ""

    for i in range(n):

        x_clock = int(X[8])
        y_clock = int(Y[10])
        z_clock = int(Z[10])


        total = x_clock + y_clock + z_clock

        if total >= 2:
            majority = 1
        else:
            majority = 0


        if x_clock == majority:
            X = clock_X(X)

        if y_clock == majority:
            Y = clock_Y(Y)

        if z_clock == majority:
            Z = clock_Z(Z)

        output_bit = int(X[-1]) ^ int(Y[-1]) ^ int(Z[-1])
        keystream += str(output_bit)

    return keystream


def clock_X(X):

    feedback = int(X[13]) ^ int(X[16]) ^ int(X[17]) ^ int(X[18])
    new_X = [str(feedback)]

    for i in range(len(X) - 1):
        new_X.append(X[i])

    return new_X


def clock_Y(Y):
    feedback = int(Y[20]) ^ int(Y[21])

    new_Y = [str(feedback)]

    for i in range(len(Y) - 1):
        new_Y.append(Y[i])
    return new_Y


def clock_Z(Z):
    feedback = int(Z[7]) ^ int(Z[20]) ^ int(Z[21]) ^ int(Z[22])

    new_Z = [str(feedback)]

    for i in range(len(Z) - 1):
        new_Z.append(Z[i])

    return new_Z


def text_to_binary(text):
    binary_result = ""

    for c in text:
        binary_char = format(ord(c), '08b')
        binary_result += binary_char

    return binary_result


def binary_to_text(binary):
    text_result = ""

    for i in range(0, len(binary), 8):
        byte = binary[i:i+8]

        text_result += chr(int(byte, 2))

    return text_result


def encrypt_A51(plaintext, X, Y, Z):

    plaintext_binary = text_to_binary(plaintext)
    keystream = A51(X, Y, Z, len(plaintext_binary))

    ciphertext_binary = ""

    for i in range(len(plaintext_binary)):
        bit_plain = int(plaintext_binary[i])
        bit_key = int(keystream[i])

        bit_cipher = bit_plain ^ bit_key
        ciphertext_binary += str(bit_cipher)

    return plaintext_binary, keystream, ciphertext_binary
```

```python
X = "1110001100101001011"
Y = "0011000000010000001101"
Z = "10011101101111001001110"



plaintexts = [
    "It is alive",
    "Snap out of it",
    "I am as mad as hell and I am not going to take this anymore",
    "Bond James Bond",
    "Love means never having to say you're sorry"
]


for plaintext in plaintexts:
    plaintext_bin, keystream, ciphertext_bin = encrypt_A51(plaintext, X, Y, Z)

    print(f"Plaintext: {plaintext}")
    print(f"Plaintext (Binary): {plaintext_bin}")
    print(f"Keystream: {keystream}")
    print(f"Ciphertext (Binary): {ciphertext_bin}")


    decrypted_bin = format(int(ciphertext_bin, 2) ^ int(keystream, 2), f'0{len(ciphertext_bin)}b')

    decrypted_text = binary_to_text(decrypted_bin)
    print(f"Decrypted: {decrypted_text}")
    print("=========================================")
```
