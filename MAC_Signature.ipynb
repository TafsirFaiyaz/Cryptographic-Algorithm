```python
import os
import sys
import binascii

from cryptography.hazmat.primitives import cmac
from cryptography.hazmat.decrepit.ciphers import algorithms

messages = [
    "I met an interesting turtle while the song on the radio blasted away",
    "I like to leave work after my eight-hour tea-break",
    "Her daily goal was to improve on yesterday",
    "He found the chocolate covered roaches quite tasty",
    "After fighting off the alligator, Brian still had to face the anaconda",
    "He decided to count all the sand on the beach as a hobby",
    "The sign said there was road work ahead so he decided to speed up",
    "Send 500$ to this account - 6589415651548",
    "Garlic ice-cream was her favorite",
    "I'd rather be a bird than a fish"
]

keys = [
    b'\x01\xd8i\xa1^0\x9a<\x0f\xf0\r\xc1\xdd\xd5\x89\xa6',
    b'\xa6+\x16\x9d-1\xda\x8aV\xed\xf5\xf0cv\x04\x88',
    b'[\xc5\xbd\xe4z\xd1=E\x17-ku\x02=|=',
    b'5"k\xff\x81a\x9b 7\x8c>\xb7\xb9\xdcu\xaa',
    b'\xa1\xfcw"?3\x91\x1c\t\x9c\x91\xe2He\x935',
    b'\xa7\x83@\xde\xbf\xb494\xee\x84\x1e-\xc8A\xf9:',
    b'2\xcbv\xdcU6\x99\xb6.\xa7\xea\xeb\xaf\x10\xc7\x90',
    b'\xc3\xea\x99e\xaal\xab\xd4\x9b\xf9\xb4Z\x19\xed\xcf\xcb',
    b'\x05\xf9\x83\x9d\xb7\xb6\xc3\xb8\x9e\xc5\xd9\xd8\x07]\xc6\xb3',
    b'\x84YY\xf0\x02GU\xa4LD\xd5\x85!A\xc2c'
]

mac_signatures = [
    "ba4ecb8db45c6ae0",
    "f47e78c537fa1435",
    "ddaf3152edbe868a",
    "9d30d856f84489a8",
    "b9d173e05bbf7738",
    "6355e471bd9930a1",
    "9fbafc75e0a5056a",
    "35273149636aca35",
    "dc2de1e07b71d391",
    "5e191d02aa5fc0b1"
]

for i in range(len(messages)):
    message = messages[i]
    message_bytes = message.encode()  # Convert string to bytes
    key = keys[i]
    # You code here. HINT: use CMAC, TripleDES
    c = cmac.CMAC(algorithms.TripleDES(key))

    c.update(message_bytes)

    signature = c.finalize()

    print(f"Message: {message}")
    print(f"CMAC signature: {binascii.b2a_hex(signature).decode()}")

    print("======================")
    given_signature = mac_signatures[i]
    if signature.hex() == given_signature:
        print("Valid")
    else:
        print("Invalid")

    print("********************************")
    print()
```
